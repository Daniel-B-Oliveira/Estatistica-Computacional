```python
import pandas as pd
import scipy as scipy
from scipy import stats
import numpy as np
```

```python
passagens = pd.read_csv("PassagensAereas.txt",sep="\t")
passagens = passagens.dropna()
passagens.head()
```

```python
passagens['price'] = pd.to_numeric(passagens['price'], errors="coerce")
passagens.head()
```

```python
passagens['destination_city'].unique()
```

```python
passagens['source_city'].unique()
```

```python
dados = passagens[passagens['source_city'] == 'Delhi']
dados = dados[dados['destination_city'] == 'Mumbai']
```

```python
dados['class'].unique()
```

```python
economica = dados[dados['class'] =='Economy']['price']
executiva = dados[dados['class'] == 'Business']['price']
```

```python
display(economica.mean())
display(executiva.mean())
```

```python
stats.kruskal(economica, executiva)
```

```python
stats.mannwhitneyu(economica, executiva)
```
