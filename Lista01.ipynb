#c) Apresente a media, mediana, valor mınimo e valor maximo, variancia, desvio padrao e coeficiente de variacao para todas as variaveis quantitativas.

```python
df_num = df.select_dtypes(exclude=["object"])
display(df_num.describe())
display(df_num.var())
```

#d) Apresente a media, mediana, valor mınimo e valor maximo, variancia, desvio padrao e coeficiente de variacao para todas as variaveis quantitativas, por turma. As turmas parecem similar quanto a media em todas as variaveis? E quanto a dispersao? Justifique.

```python
for colunas in df_num.columns:
  print(colunas)
  display(df.groupby("Turma")[colunas].describe())

df_num_2 = df.select_dtypes(exclude=["object"])
df_num_2['Turma'] = df['Turma']
print("Var")
df_num_2.groupby("Turma").var()
```

# f) Descreva todas as variaveis por meio de graficos observando o tipo de grafico mais adequado para cada uma.

```python
for coluna in df.columns:
  if df[coluna].dtype == "float" or coluna in [
      "Filhos", "Exerc", "TV", "Cine", "Idade"]:
    grafico = px.histogram(df_num, coluna, title=coluna)
    grafico.show()
  else:
    fig = px.bar(df[coluna])
    fig.show()
```

#g) Descreva todas as variaveis pormeio de graficos observando o tipo de grafico mais adequado para cada uma, por turma. Opadrao das distribuicoes parece similar nas duas turmas? Justifique.

Não, conforme é visto no gráficos.

```python
for coluna in df.columns:
  if df[coluna].dtype == "float" or coluna in [
      "Filhos", "Exerc", "TV", "Cine", "Idade"]:
    grafico = px.histogram(df, x=coluna, title=coluna, color="Turma")
    grafico.show()
  else:
    fig = px.bar(df,x=coluna, color='Turma')
    fig.show()
```
