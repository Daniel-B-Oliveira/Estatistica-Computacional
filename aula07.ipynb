# Teste para duas mdias com varincias populacionais consideradas diferentes

```python
usina = pd.read_csv("/content/Usinagem.eixos.txt", sep="\t", decimal=",")
usina.head()
```

```python
x= usina[usina['Sist-Usinagem'] == 1]['Medida']
y= usina[usina['Sist-Usinagem'] == 2]['Medida']

[stats.shapiro(x), stats.shapiro(y)]
```

#F-test

```python
def f_test(group1, group2):

  #ddof = 1: varincia amostral
  #ddof = 0: varincia populacional

  #https://numpy.org/devdocs/reference/generated/numpy.var.html#numpy.var

  f = np.var(group1, ddof=1)/np.var(group2, ddof=1)
  nun = group1.size - 1
  dun = group2.size - 1
  p_value = 2*min([1-stats.f.cdf(f, nun, dun), stats.f.cdf(f, nun, dun)])
  return f, p_value
```

```python
f_test(x,y)
#Existem ividencias para rejeitar que as varincias so iguais
```

```python
stats.ttest_ind(x, y, equal_var=False)
```
