```python
dados = pd.read_csv("/content/URSOS2.csv", sep=";")
display(dados)
```

```python
dados['IDADE'] = pd.to_numeric(dados['IDADE'], errors="coerce")
grafico = px.histogram(dados, x='IDADE', color="SEXO")
grafico.show()
```

```python
dados['FAIXAETARIA'] = pd.to_numeric(dados['FAIXAETARIA'], errors="coerce")
grafico = px.histogram(dados, x='FAIXAETARIA', color="SEXO")
grafico2 = px.histogram(dados, x='FAIXAETARIA')
grafico2.show()
grafico.show()
```

```python
grafico = px.histogram(dados, x='MES', color='SEXO')
grafico.show()
```

```python
dados.columns
exercicio_d = ['COMPCAB', 'LARGCAB', 'PESC','COMPCOR', 'PEITO', 'PESO']
for colunas in exercicio_d:
  dados[colunas] = pd.to_numeric(dados[colunas], errors="coerce")
  grafico = px.histogram(dados, x = colunas, color='SEXO')
  grafico.show()
```

```python
exercicio_5 = ['IDADE','COMPCAB', 'LARGCAB', 'PESC','COMPCOR', 'PEITO', 'PESO']

display(dados.groupby('SEXO')[exercicio_5].mean())
```
